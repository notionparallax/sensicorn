```python
from scipy.misc import imread
from StringIO import StringIO
import calendar
import datetime
import dateutil.parser
import json
import math
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd
import requests
```

```python
#get the base data, just to get the names of the bases out of it
def getDFfromCSVURL(url,columnNames=False):
    r = requests.get(url)
    data = r.content
    if columnNames:
        return pd.read_csv(StringIO(data),header=0,names=columnNames)
    else:
        return pd.read_csv(StringIO(data))


baseData = getDFfromCSVURL("https://docs.google.com/spreadsheets/d/167pxh_NRq5e9sQV9Zb8Z-Q-kdN7Zy9FCzCp47eVXTuE/pub?gid=1446311565&single=true&output=csv")

#only keep the columns we care about
# print "all columns: ", baseData.columns.values
# ['agentId' 'agentName' 'location' 'x' 'y' 'HANG OR MOUNT?' 'WHERE AND HOW IS IT POWERED?' 'ADDITIONAL EQUIP.'
#  'ACTIVE OR INACTIVE?' 'MIGHT NEED +WIFI' 'ON MONITORS' 'BASESTATION LAYOUT']
baseData = baseData.ix[:,0:3]
baseData = baseData[baseData.agentId > 0] #not really sure how this works, maybe failure is falsy?
baseData.head(3)
```

```python
with open("toilet.json") as f:
   data1 = f.read()
with open("othertoilet.json") as f:
   data2 = f.read()
    
# Raw from the server, the timestamp isn't valid json.
# Run this regex: ISODate\((\"[\d-]+T[\d\.:]+Z\")\) and replace with $1 to fix it
# there are also 2 lines at the top
```

```python
df1 = pd.read_json(data1)
df2 = pd.read_json(data2)
```

```python
df = pd.concat([df1,df2])
df = df.merge(baseData,on="agentId")
df['pTime'] = df.apply(lambda x: dateutil.parser.parse(x.time) , axis=1) 
df = df.drop(['time'], 1)
df.rename(columns={"pTime": "time"}, inplace=True)
print df.shape
df.head()
```

```python
with open("agentUptime.json") as f:
   updata = f.read()
    
utdf = pd.read_json(updata)

utdf['pTime'] = utdf.apply(lambda x: dateutil.parser.parse(x.time) , axis=1)
utdf = utdf.drop(['time'], 1)
utdf.rename(columns={"pTime": "time"}, inplace=True)

utdf["dayNum"   ]= utdf.apply(lambda x:                   x.time.weekday()  , axis=1)
# utdf["dayName"  ]= utdf.apply(lambda x: calendar.day_name[x.time.weekday()] , axis=1)
# utdf["justTime" ]= utdf.apply(lambda x:                   x.time.time()     , axis=1)
utdf["dayOfYear"]= utdf.apply(lambda x:                   dayOfYear(x)      , axis=1)
utdf["hour"     ]= utdf.apply(lambda x:                   x.time.hour       , axis=1)
utdf = utdf[utdf.time>df.time.min()]
utdf.head()
```
