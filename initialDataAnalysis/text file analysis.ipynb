```python
import datetime
import dateutil.parser
import json
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd
import requests
from scipy.misc import imread
from StringIO import StringIO
```

```python
%matplotlib inline
plt.rcParams['figure.figsize'] = (20, 10)
```

```python
#get the data about the base stations from the google spreadsheet
baseDataCSVaddress = "https://docs.google.com/spreadsheets/d/167pxh_NRq5e9sQV9Zb8Z-Q-kdN7Zy9FCzCp47eVXTuE/pub?gid=1446311565&single=true&output=csv"
r = requests.get(baseDataCSVaddress)
data = r.content
baseData = pd.read_csv(StringIO(data))

#only keep the columns we care about
baseData = baseData.ix[:,0:5]
baseData = baseData[baseData.agentId > 0] #not really sure how this works, maybe failure is falsy?
baseData.head(3)
```

```python
# read the entire file into a python array
thisManyRowsOfData = 200000
d = []
with open("200916_2_3.json", 'rb') as f:
    for i in range(thisManyRowsOfData):
        d.append(f.readline())
```

The data comes in as some kind of skanky hybrid of json and, er, something else. This section is ripped from SO and fixes it

```python
# remove the trailing "\n" from each line
data = map(lambda x: x.rstrip(), d)

# each element of 'data' is an individual JSON object.
# i want to convert it into an *array* of JSON objects
# which, in and of itself, is one large JSON object
# basically... add square brackets to the beginning
# and end, and have all the individual business JSON objects
# separated by a comma
data_json_str = "[" + ','.join(data) + "]"

# now, load it into pandas
df = pd.read_json(data_json_str)

#get rid of the mongo ID. We don't need it.
del df['_id']

#convert the mongo time stamp into a python datetime object 
def toDatetime(mongoDate):
    d=datetime.datetime.fromtimestamp(mongoDate["time"][u'$date']/1000.0) #reported in ms, needs to be s
    return d
    
df["timestamp"] = df.apply(toDatetime, axis=1)

#get rid of the mongo time. We don't need it.
del df['time']
```

```python
#This gives us an idea of the time domain that the data we have is interested in.
def timeStats(df, showprint=True):
    minT = df.timestamp.min()
    maxT = df.timestamp.max()
    rangeT = maxT-minT
    if showprint:
        print minT,   "First Recording in this set"
        print maxT,   "Last Recording in this set"
        print rangeT, "Range covered"
    return (minT,maxT,rangeT)

timeStats(df)
```

This is all very well, but how do we tell what happened over time. This is 3 minutes of data. Were they moving or is there just a massive detection radius?

Lets try to animate things:

```python
#this takes quite a long time
def timeSliced(startTime, windowLength, df):
    if type(startTime) is str:
        startTime = dateutil.parser.parse(startTime)
    delta = datetime.timedelta(seconds=windowLength)
    endTime = startTime+delta
    df = df[(df.timestamp>startTime) & 
            (df.timestamp<endTime)]
    centroid = {"x":df["x"].mean(), "y":df["y"].mean()}
    #print "start:",startTime," end:",endTime, " delta:",delta, df.shape#, "\n",df.timestamp
    
    plt.title('Detections for person 297 from '+str(startTime)+" to "+str(startTime+delta)) #TODO, make this a bit nicer
    img = imread("map.png")
    
    plt.scatter(df.x, df.y, zorder=1, s=abs(df.rssiAdj)*100, alpha=0.2)
    
    plt.scatter(centroid["x"], centroid["y"], c="r", zorder=2, s=100)
    
    plt.imshow(img, zorder=0, extent=[-1300, 72000, -24000, 41000])
    #                                [left,  right, bottom, top  ]
    # plt.show() #don't show, we need to save
    plt.savefig("anim/foo"+str(startTime)+".png", bbox_inches='tight')
    plt.clf()

ts=timeStats(person297df)
timeSize = ts[2]
stepSize = 0.5 #in seconds
steps = int(timeSize.total_seconds() / stepSize)
print "will make",steps,"frames"

for i in range(steps):
    s = ts[0]+datetime.timedelta(seconds=stepSize*i)
    timeSliced(s, 
               stepSize, 
               person297df)
#send a system command to make this into a gif. See https://community.linuxmint.com/tutorial/view/1118
os.system("convert -delay 20 -loop 0 anim/*.png result.gif")
```

This image shows the detections from person 297.

The red dot is the unweighted centroid of the detections in that window.

It will be interesting to play around with weighting the centroid by detection strength.

```python
#makes centroids for each time window, This then gets shown on a static image.
def timeSlicedCentroids(startTime, windowLength, df):
    if type(startTime) is str:
        startTime = dateutil.parser.parse(startTime)
    delta = datetime.timedelta(seconds=windowLength)
    endTime = startTime+delta
    df = df[(df.timestamp>startTime) & 
            (df.timestamp<endTime)]
    #print "start:",startTime," end:",endTime, " delta:",delta, df.shape#, "\n",df.timestamp
    return {"x":df["x"].mean(), "y":df["y"].mean()}

centroids = []
for i in range(steps):
    s = ts[0]+datetime.timedelta(seconds=stepSize*i)
    tsc = timeSlicedCentroids(s, 
               stepSize, 
               person297df)
    centroids.append(tsc)

centroidsdf = pd.DataFrame(centroids)
centroidsdf.dropna(inplace=True)
centroidsdf.head()
```

```python
plt.title('Detections and centroids for person 297')
img = imread("map.png")
plt.scatter(person297df.x, person297df.y, zorder=1, s=abs(person297df.rssiAdj)*100, alpha=0.1)
plt.plot(   centroidsdf.x, centroidsdf.y, c="r", zorder=2)
plt.scatter(centroidsdf.x, centroidsdf.y, c="g", zorder=3, s=30)
plt.imshow(img, zorder=0, extent=[-1300, 72000, -24000, 41000])
#                                [left,  right, bottom, top  ]
plt.show()
print "All detections in this time period. Circle radius shows signal strength."
print "Time window size:",stepSize,"seconds"
timeStats(person297df)
```

If we try a different approach, what does it look like from the point of view of the base station?

This graph shows the rssi for the bases that detected this person. 

```python
basesToCareAbout = person297df.agentName.unique()

fig, ax = plt.subplots()
plt.title('Detection strengths for person 297 per base station')
legendStuff = []
for base in basesToCareAbout:
    localDF = person297df[person297df.agentName == base]
    plt.plot_date(localDF.timestamp, localDF.rssi, ls='solid')
    legendStuff.append(base)
plt.legend(legendStuff)
start, end = ax.get_xlim() #set these so that the individual plots don't bounce
bottom, top = ax.get_ylim()
step = (end-start)/30
ax.xaxis.set_ticks(np.arange(start, end, step))
dateFormat = matplotlib.dates.DateFormatter('%M:%S')
ax.xaxis.set_major_formatter(dateFormat)
plt.show()
```

```python
basesToCareAbout = person297df.agentName.unique()

for base in basesToCareAbout:
    plt.title('Detection strengths for person 297 per base station')
    localDF = person297df[person297df.agentName == base]
    plt.plot_date(localDF.timestamp, localDF.rssi, ls='solid')
    plt.legend([base])
    plt.xlim([start, end])
    plt.ylim([bottom, top])
    plt.savefig("lineplt/lines"+base+".png", bbox_inches='tight')
    plt.clf()

#send a system command to make this into a gif. See https://community.linuxmint.com/tutorial/view/1118
os.system("convert -delay 20 -loop 0 lineplt/*.png linePlots.gif")
```
